# 2. Mutate

```python
from GAComposition import *

GAAlg = GAComposition(scoreOriginal)
GAAlg.evolve(5)
top3 = GAAlg.getTop3()
```

```python
m21Part_main = measures_to_m21Part(top3[1].measures)

#build
title = name + ' GAmut'
score_GA = build_m21Score_1p(m21Part_main,title,scoreOriginal.ts)
#score_GA.show('lily.png')
score_GA.show()
```

```python
score_GA.show('midi')
```
